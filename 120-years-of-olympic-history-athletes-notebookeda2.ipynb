```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.displot(veri,x='kilo')
plt.ylabel("frekans")
plt.title("Arlk Histogram")
plt.show()
```

```python
sns.displot(veri,x="kilo",hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete gre ayrtrlm arlk histogram")
```

```python
sns.displot(veri,x="kilo",y="boy",kind="kde")
plt.xlabel("Arlk")
plt.ylabel("Boy")
plt.title("Arlk boy histogram")
plt.show()
```

```python
sns.kdeplot(data=veri,x="kilo",y="boy",hue="cinsiyet",fill=True)
plt.xlabel("Arlk")
plt.ylabel("Boy")
plt.title("Cinsiyete gre ayrtrlm Arlk boy histogram")
plt.show()
```

```python
g=sns.JointGrid(data=veri,x="kilo",y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
sns.pairplot(veri)
```

```python
g=sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
```
